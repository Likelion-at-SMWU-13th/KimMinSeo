## 01. mpg.csg를 불러와서 mpg.csv에서 displ이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy 평균이 더 높은지 알아보기 ##

### 1-1. mpg.csv 불러오기 ###

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg
```

*→ 234개의 행과 11개의 열(변수)로 이루어진 데이터 생성*

### 1-2. 조건에 해당하는 데이터 추출하여 데이터 만들기

```python
displ4 = mpg.query('displ <= 4')
displ5 = mpg.query('displ >= 5')
```

```python
displ4
```

```python
displ5
```

*→ 배기량이 4이하인 자동차, 5이상인 자동차의 데이터를 따로 생성*

### 1-3. 각 데이터의 hwy 평균 확인 ### 

```python
displ4['hwy'].mean()
```

```python
displ5['hwy'].mean()
```

*→ displ4 hwy평균은 약 25.96이고  displ5의 hwy평균은 약 18.08이므로 배기량 4이하 자동차의 고속도로 연비 평균이 더 높다.*

## 02. audi와 toyota 중 어느 manufactuer의 city 평균이 높은지 알아보기 ##

### 2-1. 조건 해당하는 행 추출하여 데이터 만들기 ### 

```python
audi = mpg.query('manufacturer == "audi"')
toyota = mpg.query('manufacturer == "toyota"')
```

```python
audi
```

```python
toyota
```

*→ 자동차 제조회사가 audi인 자동차, toyota인 자동차의 데이터를 따로 생성*

### 2-2. 각 데이터의 cty 평균 확인 ###

```python
audi['cty'].mean()
```

```python
toyota['cty'].mean()
```

*→ audi의 cty평균은 약 17.6이고 toyota의 cty평균은 약 18.5이므로 자동차 제조회사가 toyota인 자동차들의 도시 연비 평균이 더 높다.*
