```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
```

```python
exam.assign(total = exam['front']+exam['back']+exam['design'])
```

```python
exam.assign(total = exam['front']+exam['back']+exam['design'],
            mean = (exam['front']+ exam['back']+exam['design'])/3)
```

```python
exam.assign(test = np.where(exam['design']>=60, 'pass', 'fail'))
```

```python
exam.assign(total = exam['front']+exam['back']+exam['design']) \
    .sort_values('total')
```

```python
exam.assign(total = exam['front']+exam['back']+exam['design'], 
            mean = lambda x: x['total']/3)
```

```python
exam.agg(mean_front = ('front', 'mean'))
```

```python
exam.groupby('nclass') \
    .agg(mean_front = ('front', 'mean'))
```

```python
exam.groupby('nclass') \
    .agg(mean_front = ('front', 'mean'),
         sum_front = ('front', 'sum'),
         median_front = ('front', 'median'),
         n = ('nclass', 'count'))
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.query('category == "suv"') \
    .assign(total = (mpg['hwy']+mpg['cty'])/2) \
    .groupby("manufacturer") \
    .agg(mean_tot = ('total', 'mean')) \
    .sort_values('mean_tot', ascending = False) \
    .head()
```
