```python
import pandas as pd

df_midwest = pd.read_csv('midwest.csv')
df_midwest
```

## 명령어 사용하여 데이터 특징 파악하기 ##
메서드: 변수가 지니고 있는 함수로, 변수 뒤에 점을 찍어 사용한다. 변수의 자료 구조에 따라 사용가능 메서드가 다르다.

### head(), tail()로 앞뒤 행의 데이터 출력 ###


```python
df_midwest.head()
```

```python
df_midwest.head(10)
```

```python
df_midwest.tail()
```

```python
df_midwest.tail(10)
```

### shape로 데이터프레임의 행열 개수 출력 ###
shape는 함수가 아닌 속성(attribute)이므로 괄호를 입력하지 않는다!

```python
df_midwest.shape
```

*-> df_midwest가 437개의 행, 28개의 열을 가지고 있음을 확인할 수 있다.*

### info()로 모든 변수의 속성 한눈에 파악하기 ###
info()에도 행열 개수 정보가 출력된다. 한눈에 파악하기 위해 shape보다 더 자주 사용되는 명령어어

```python
df_midwest.info()
```


- *첫째 줄에서 df_midwest가 padas로 만들어진 데이터프레임이라는 것을 확인*
- *둘째 줄에서 437개의 행으로 이루어졌으며, 행 번호가 0부터 436까지라는 것을 확인 (cf. 파이썬의 셈은 0부터 시작)*
- *셋째 줄에서 df_midwest가 변수(열) 28개로 구성된다는 것을 확인*
- *이하에서 df_midwest의 변수 속성을 확인 (county, state, category를 제외하면 전부 float/int 계량형)*

### describe()로 변수 값의 요약통계량 확인하기 ###

```python
df_midwest.describe()
```

```python
df_midwest.describe(include= 'all')
```

- *문자인 변수의 요약 통계량을 출력하려면 .describe(include = 'all')*
- *unique, top, freq는 문자 변수로만 계산. 숫자 변수에는 NaN*
